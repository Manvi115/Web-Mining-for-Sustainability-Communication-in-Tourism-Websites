```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from collections import Counter
import pickle
```

```python
base_url = 'https://arunachaltourism.com'  # Simple website for testing
response = requests.get(base_url)
soup = BeautifulSoup(response.text, 'html.parser')

# Extract and print all links
links = []
for a_tag in soup.find_all('a', href=True):
    link = a_tag['href']
    full_link = urljoin(base_url, link)
    if full_link.startswith(base_url):
        links.append(full_link)

print(base_url, "links are",links)
```

```python

all_words = []

# Now, scrape each of those links
for link in links:
    try:
        # Fetch the content of each linked page
        response = requests.get(link)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract text including hyperlinks (anchor text)
        text = soup.get_text(separator=" ")  # Using a space as separator for cleaner words

        # Also extract anchor text (hyperlink text)
        anchor_texts = [a.get_text() for a in soup.find_all('a')]

        # Combine the two sources of text
        full_text = text + " " + " ".join(anchor_texts)

        # Split into words and add to the list
        words = full_text.split()
        all_words.extend(words)

    except Exception as e:
        print(f"Failed to scrape {link}: {e}")

# Save all the words to a Pickle file
with open('lexicons.pkl', 'wb') as f:
    pickle.dump(all_words, f)

# Create a Counter to count the occurrences of each word (including duplicates)
word_counts = Counter(all_words)
```

```python
# Example: Search for a word and display how many times it was found
word_to_search = "UNESCO"  # The word you want to search for
count = word_counts.get(word_to_search, 0)

if count > 0:
    print(f"The word '{word_to_search}' was found {count} times.")
else:
    print(f"The word '{word_to_search}' was NOT found.")
```
