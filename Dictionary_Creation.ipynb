```python
import kagglehub
import os
from gensim.models import KeyedVectors

# Download latest version
dataset_path = kagglehub.dataset_download("leadbest/googlenewsvectorsnegative300")

# Find the model file within the downloaded directory - assuming it's named 'GoogleNews-vectors-negative300.bin'
#  You might need to adjust this if the filename is different
model_filename = "GoogleNews-vectors-negative300.bin"
path = os.path.join(dataset_path, model_filename)

print("Path to dataset files:", path)

# Load pre-trained Word2Vec model
model = KeyedVectors.load_word2vec_format(path, binary=True)

# Function to find related words
def get_related_words(word, top_n=10):
    return [w[0] for w in model.most_similar(word, topn=top_n)]
```

```python
# Function to add new words
def add_words_to_dict(category, new_words):
    if category in sustainability_dict:
        sustainability_dict[category].extend(new_words)
    else:
        # If category doesn't exist, create a new one
        sustainability_dict[category] = new_words
```

```python
# Expand each key in the dictionary with related words
for key in sustainability_dict:
    related_words = get_related_words(key.split()[0].lower())  # Use the first word of each key (e.g., "nature", "culture")
    add_words_to_dict(key, related_words)
    print(f"{key} Related Words: {related_words}")
    print(f"{key} updated")
```
