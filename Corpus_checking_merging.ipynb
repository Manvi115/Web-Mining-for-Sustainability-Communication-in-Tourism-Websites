```python
import os
import pickle

# Path to the folder containing your .pkl files
folder_path = '/content/drive/MyDrive/WM_Project/Corpus_states'

# List all files in the directory
file_list = [f for f in os.listdir(folder_path) if f.endswith('.pkl')]

# Initialize an empty list to hold the merged corpus
merged_corpus = []

# Loop through all .pkl files and merge them
for file_name in file_list:
    file_path = os.path.join(folder_path, file_name)

    # Load each .pkl file
    with open(file_path, 'rb') as file:
        corpus = pickle.load(file)

        # Ensure the corpus is not empty before merging
        if corpus:
            merged_corpus.extend(corpus)
        else:
            print(f"Warning: {file_name} is empty or contains no data.")

# Save the merged corpus back to the folder
merged_corpus_path = os.path.join(folder_path, 'merged_corpus.pkl')

with open(merged_corpus_path, 'wb') as file:
    pickle.dump(merged_corpus, file)

print(f'Merged corpus saved to {merged_corpus_path}. Total documents: {len(merged_corpus)}')
```
